# Projeto Banco 2º nível

```python
'''
    Projeto conta bancária
        2º Desafio
'''

menu = '''
===========Banco==========
|   [u] Criar usuário    |
|   [c] Conta corrente   |
|   [lc]Listas de contas |
|   [d] Deposito         |
|   [s] Saque            |
|   [e] Extrato          |
|   [q] Sair             |
==========================
'''

AGENCIA = '0001'
saque_limite = 3
saldo = 0
valor_maximo_por_saque = 500
contador_de_saques = 0
extrato = ""
cont_deposito = 0
valor_atual = 0
saques = 0
extrato2 = ""
usuarios = []
contas = []

def listar_contas(contas):
    for conta in contas:
        linha = f'''
        Agência:\t{conta['agencia']}
        C\C:\t\t{conta['numero_da_conta']}
        Titular:\t{conta['usuario']['nome']}
        '''
    print(linha)
    print("="*100)
        
def filtrar_usuario(cpf,usuarios):
    usuarios_filtrados = [usuario for usuario in usuarios if usuario['cpf'] == cpf]
    return usuarios_filtrados[0] if usuarios_filtrados else None

def criar_usuario(usuarios):
    cpf = input('Digite seu CPF: ')
    usuario = filtrar_usuario(cpf,usuarios)
    
    if usuario:
        print('Já há um usuario utilizando este cpf')
        
    nome = input("Escreva seu nome completo: ")
    data_de_nascimento = input('Insira sua data de nascimentp(dd-mm-aaaa): ')
    endereco = input('Logradouro - nro - bairro - cidade/sigla estado: \n') 
    usuarios.append({'nome':nome,'data_de_nascimento':data_de_nascimento ,'cpf':cpf,'endereço':endereco})
        
def conta_corrente(AGENCIA,numero_da_conta,usuarios): 
    cpf = input('Digite seu CPF: ')
    usuario = filtrar_usuario(cpf,usuarios)
    if usuario:
        print('====Conta criada com sucesso====')
        return {"agencia":AGENCIA,'numero_da_conta':numero_da_conta,'usuario':usuario}
    print('Usuário não encontrado')
    
def sacar(s):
    global contador_de_saques
    global valor_atual
    global saldo
    global extrato
    
    if s <= saldo:
        if contador_de_saques < saque_limite:
            if s <= 500:
                print('Transação realizada com sucesso ') 
                contador_de_saques += 1
                valor_atual = saldo - s
                saldo=valor_atual
                extrato += f'R$ {s:.2f}\n'
                #print(saldo)
            else:
                print('''Você excedeu o valor maximo permitido de saque,\n ensira outro valor menor ou igual a 500 reais''')
        else:
            print('Você chegou ao seu limite de saques diários')
        print('\n')
    else:
        print("Você não tem saldo suficiente, tente outro valor")

def deposito(d):
    global saldo
    global cont_deposito
    global extrato2
    
    saldo += d
    cont_deposito += 1
    extrato2 += f'R$ {d:.2f}\n'
    print('\n')
    
def historico():
    print(f'''
===============Extrato===============
Usúario: \n{usuarios[0]['nome']}
Quantidade de saques: {contador_de_saques}
Quantidade de depositos: {cont_deposito}
Deposito em R$: \n{extrato2}
Saque em R$: \n{extrato}
=====================================
''')       

#Obs.: Me lembrar caso seja necessario
'''
CPF: {usuarios[0]['cpf']}
Saldo atual: {saldo:.2f}
'''

while True:
    print(menu)
    choose = input('Escolha sua opção:')
    print('\n')
    if choose == 'u':
        criar_usuario(usuarios=usuarios)
    elif choose == 'd':
        d = float(input('Digite o valor que deseja depositar: '))
        deposito(d)
    elif choose == 'c':
        numero_da_conta = len(contas) + 1
        conta = conta_corrente(AGENCIA,numero_da_conta,usuarios)
        if conta:
            contas.append(conta)
    elif choose == 'lc':
        listar_contas(contas)
    elif choose == 's':
        s = float(input('Digite o valor que deseja sacar: '))
        sacar(s=s)
    elif choose == 'e':
        historico()
    elif choose == 'q':
        break
    else:
        print("Tente novamente, insira uma opção existente.")
        print('\n')  
print('Obrigado por escolher nosso banco, estamos preparados para lhe ajudar no quer for preciso!!')
```
